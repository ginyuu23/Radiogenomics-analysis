```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/Users/armandayuu/Desktop/1/exp0630/spearman111_plot.csv" ,header=0, index_col=0)
#t_data = pd.DataFrame(data)
data1 = abs(data)
data1.head()
```

```python
#输出热力图
plt.figure(figsize=(6,8))
out = sns.set (font_scale=1.1)
out = sns.heatmap(data1, vmin=0.4, vmax=0.56, annot=True,annot_kws={'size':13,'weight':'bold'}, cmap= "RdBu_r")

#保存热力图
output = out.get_figure()

plt.show()
```

```python
#保存热力图
output.savefig('/Users/armandayuu/Desktop/HeatMap111.png',dpi=800,bbox_inches='tight')
```
